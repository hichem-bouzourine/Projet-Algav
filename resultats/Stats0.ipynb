```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

# Tas Min 

## Ajouts Itératifs

```python
# Lecture des données depuis le fichier CSV 
#!  Toute les clés sont utilisés
df_ajout_iteratif_arbre = pd.read_csv('./ajout_iteratifs_tas_arbre.csv')
df_ajout_iteratif_tab = pd.read_csv('./ajout_iteratifs_tas_tab.csv')
```

```python
df_ajout_iteratif_arbre['moyenne']
```

```python
# Extraire les colonnes nécessaires
taille_ajout_iteratif_arbre = df_ajout_iteratif_arbre['Taille']
moyenne_ajout_iteratif_arbre = df_ajout_iteratif_arbre['moyenne']
taille_ajout_iteratif_tab = df_ajout_iteratif_tab['Taille']
moyenne_ajout_iteratif_tab = df_ajout_iteratif_tab['moyenne']

# Créer le graphique
# Créer le graphique
plt.figure(figsize=(10, 6))

# Tracer les données pour l'arbre
plt.plot(taille_ajout_iteratif_arbre, moyenne_ajout_iteratif_arbre,  label='Arbre')

# Tracer les données pour le tableau
plt.plot(taille_ajout_iteratif_tab, moyenne_ajout_iteratif_tab,  label='Tableau')

# Ajouter des labels et un titre
plt.title('Temps d\'ajout itératifs en fonction de la taille de la structure')
plt.xlabel('Taille de la structure')
plt.ylabel('Temps moyen d\'ajout (ms)')

# Ajouter une légende
plt.legend()

# Afficher le graphique
plt.grid(True)
plt.show()
```

# Comparaison - Shakespear

## Ajout 

```python
# Lecture des données depuis le fichier CSV
df_ajout_tas_tab_vs_fileBinomiale = pd.read_csv('./ajout_tas_tab_vs_fileBinomiale_shakespeare.csv')
```

```python
df_ajout_tas_tab_vs_fileBinomiale['Taille']
```

```python
# Extraire les colonnes nécessaires
taille = df_ajout_tas_tab_vs_fileBinomiale['Taille']
moyenne_ajout_tas_tab = df_ajout_tas_tab_vs_fileBinomiale['moyenne_tas']
moyenne_ajout_file = df_ajout_tas_tab_vs_fileBinomiale['moyenne_file']

# Créer le graphique
plt.figure(figsize=(10, 6))
plt.plot(taille, moyenne_ajout_tas_tab, marker='o' , label="Tas (Tableau)")
plt.plot(taille, moyenne_ajout_file, marker='o', label="File")

# Ajouter des labels et un titre
plt.title('Temps nécessaire pour l\'ajout d\'un élément dans une file binomiale VS Tas min (tab) en fonction de la taille de la file')
plt.xlabel('Taille de la file (nombre de clés)')
plt.ylabel('Temps moyen d\'ajout')


plt.legend()

# Afficher le graphique
plt.grid(True)
plt.show()
```
